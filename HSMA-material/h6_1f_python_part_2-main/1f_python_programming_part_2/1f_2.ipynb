**Task 1** Write code that asks the user to input their age and stores the input as an integer.  If anything goes wrong with this, a message should be printed and the user should be asked again.  If (and only if) a valid input is obtained, the user should be informed how old they will be at their next birthday.  Regardless of whether the user input something valid or not, a message should be printed saying what a great little app this is.  If a valid input was entered, the program will then terminate, but if there was no valid input, the user will be asked again (and continued to be asked until a valid input is provided).  You should use an exception handling framework for this task.

```python
while True:
    try: 
        age = int(float(input("How old are you? ")))
        break
    except ValueError:
        print(
            f"I'm sorry, that was not a valid answer. Please enter your", 
            f"age in years as a number: "
            )
        continue
    finally:
        print("This is a great little app!")

snark = ["You are too young for this!", 
         "An early starter!", 
         "I hope you are having fun in your life!", 
         "Time to get down to some serious work now", 
         "May all your little one's troubles be GCSEs", 
         "Is it getting lonely yet?", 
         "OMG", 
         "Have you got your slippers on?", 
         "Really?", 
         "I am impressed", 
         "If that is true, that is amazing. Did you get a telegram?"
        ]
if age == 41:         
    print("You will find the answer to life, the universe, and everything.")
if age > 108:
    print(f"Shome mishtake, shurley? I don't believe you are {age}")
else:
    print(f"At your next birthday you will be {age +1} years old.", 
        f"{snark[int((age+1)/10)]}"
        )
    
```
