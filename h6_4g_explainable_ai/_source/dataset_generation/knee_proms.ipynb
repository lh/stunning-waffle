```python
import pandas as pd
```

```python
knee_proms_raw = pd.read_parquet("knee-provider.parquet")
knee_proms_raw.head()
```

```python
knee_proms_outcome_col = knee_proms_raw[['t1_eq_vas']]

knee_proms_outcome_col
```

```python
knee_proms_available_prior_surg = knee_proms_raw.drop(list(knee_proms_raw.filter(regex='t1')), axis=1)
```

```python
knee_proms_final = pd.merge(
    left=knee_proms_available_prior_surg,
    right=knee_proms_outcome_col,
    left_index=True,
    right_index=True)
```

```python
knee_proms_final.to_csv("../teaching_notebooks/knee_proms_processed.csv", index=False)
```
